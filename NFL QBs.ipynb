# Cleaning Data

```python
raw_df = pd.read_csv('NFL-QBs-2018.csv')
```

```python
df = raw_df.drop(columns = 'Rk')
df = df.set_index('Player')
```

```python
df = df[df['Pos'] == 'QB']
df = df[df['GS'] >= 10]
#Limiting our sample to Quarterbacks who started 10 games or more
```

# Finding Average Quarterback Rating of Most Accurate QBs

```python
acc_df = pd.DataFrame(raw_df['Player'])
acc_df['GS'] = raw_df['GS']
acc_df['Pos'] = raw_df['Pos']
acc_df['QBR'] = raw_df['QBR']
acc_df['Cmp%'] = raw_df['Cmp%']
acc_df['Y/A'] = raw_df['Y/A']
acc_df = acc_df[acc_df['Pos'] == 'QB']
acc_df = acc_df[acc_df['GS'] >= 10]
acc_df = acc_df.set_index('Player')
acc_df.head()
```

```python
top_acc_df = acc_df.sort_values('Cmp%', ascending = False).head(n=10)
top_acc_df
```

```python
print(top_acc_df['QBR'].mean())
print(top_acc_df['Y/A'].mean())
```
